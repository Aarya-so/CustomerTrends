```python
cities = df['location'].dropna().unique()
cities
```

```python
import requests
import time

API_KEY = "your_key"
weather_data = {}

for city in cities:
    try:
        url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={API_KEY}&units=metric"
        response = requests.get(url).json()

        weather_data[city] = {
            "temperature": response['main']['temp'],
            "humidity": response['main']['humidity'],
            "weather_condition": response['weather'][0]['main']
        }

        time.sleep(1)   # avoid API rate limit

    except:
        weather_data[city] = {
            "temperature": None,
            "humidity": None,
            "weather_condition": None
        }
```

```python
df['Temperature'] = df['location'].map(lambda x: weather_data.get(x, {}).get('temperature'))
df['Humidity'] = df['location'].map(lambda x: weather_data.get(x, {}).get('humidity'))
df['Weather'] = df['location'].map(lambda x: weather_data.get(x, {}).get('weather_condition'))
```
